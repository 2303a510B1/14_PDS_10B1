```
# 3. Writing to a file
# Open file in write mode
fp = open("example.txt", "w")

# Write text to the file
fp.write("Hello, B.Tech students!\n")
fp.write("This is file handling in ")
fp.write("SR University.\n")

# Close the file after writing
fp.close()


# 4. Reading a file
# Open file in read mode
file = open("example.txt", "r")

# Read entire content
content = file.read()
print("File Content:\n", content)

# Close file
file.close()


# 5. Reading Line by Line
file = open("example.txt", "r")

# Read first line
line1 = file.readline()
print("First Line:", line1.strip())

# Read remaining lines
line2 = file.readline()
print("Second Line:", line2.strip())

line3 = file.readline()
print("Third Line:", line3.strip())

# Close file
file.close()


# 6. Reading All Lines into a list
file = open("example.txt", "r")
lines = file.readlines()
print("All lines as List:", lines)
file.close()


# 7. Appending Data to an Existing File
file = open("example.txt", "a")
file.write("Appending new data...\n")
file.close()


# 8. Using with Statement (Best practice)
# The with statement automatically closes the file after use
with open("example.txt", "r") as file:
    data = file.read()
    print("Data using 'with' statement:\n", data)


# 9. Working with CSV Files
import csv

# Writing to CSV
with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age", "Course"])  # Header
    writer.writerow(["Sathwik", 21, "CSE"])
    writer.writerow(["Karthikeya", 21, "ECE"])
    writer.writerow(["Nithin", 21, "EEE"])

# Reading from CSV
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    print("CSV File Content:")
    for row in reader:
        print(row)
```
