```
# Install required libraries
!pip install pandas numpy matplotlib seaborn scipy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -----------------------------
# 1. Create Synthetic Dataset
# -----------------------------
np.random.seed(42)
n = 200  # number of children

data = {
    "Child_ID": range(1, n+1),
    "Age": np.random.randint(6, 16, size=n),  # school-aged kids
    "Height_cm": np.random.normal(135, 10, n),  # height
    "Weight_kg_Before": np.random.normal(35, 8, n),  # before intervention
    "Daily_ScreenTime_hrs_Before": np.round(np.random.normal(4, 1.2, n), 1),
    "Sleep_Hours_Before": np.round(np.random.normal(7, 1, n), 1),
    "Academic_Performance_Before": np.random.randint(40, 90, size=n),
    "Behavior_Score_Before": np.random.randint(1, 6, size=n)
}

df = pd.DataFrame(data)

# Calculate BMI Before
df["BMI_Before"] = df["Weight_kg_Before"] / ((df["Height_cm"]/100)**2)

# -----------------------------
# 2. Simulate After Intervention
# -----------------------------
df["Daily_ScreenTime_hrs_After"] = np.round(df["Daily_ScreenTime_hrs_Before"] * np.random.uniform(0.6, 0.7, n), 1)
df["Sleep_Hours_After"] = df["Sleep_Hours_Before"] + np.random.normal(1, 0.5, n)
df["Academic_Performance_After"] = df["Academic_Performance_Before"] + np.random.randint(5, 15, size=n)
df["Behavior_Score_After"] = np.clip(df["Behavior_Score_Before"] + np.random.randint(0, 2, size=n), 1, 5)

# Weight reduction (due to better lifestyle)
df["Weight_kg_After"] = df["Weight_kg_Before"] - np.random.normal(1, 0.5, n)
df["BMI_After"] = df["Weight_kg_After"] / ((df["Height_cm"]/100)**2)

# -----------------------------
# 3. Statistical Tests
# -----------------------------
print("\n--- Paired T-tests (Before vs After) ---")
for col in ["Daily_ScreenTime_hrs", "Sleep_Hours", "Academic_Performance", "Behavior_Score", "BMI"]:
    before = df[f"{col}_Before"]
    after = df[f"{col}_After"]
    t_stat, p_val = stats.ttest_rel(before, after)
    print(f"{col}: t={t_stat:.2f}, p={p_val:.4f}")

# -----------------------------
# 4. Correlation Analysis
# -----------------------------
corr = df[["Daily_ScreenTime_hrs_Before", "Sleep_Hours_Before",
           "Academic_Performance_Before", "Behavior_Score_Before", "BMI_Before"]].corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix (Before Intervention)")
plt.show()

# -----------------------------
# 5. Visualizations
# -----------------------------
# Scatter plot: Screen Time vs Academic Performance
plt.figure(figsize=(7,5))
sns.scatterplot(x="Daily_ScreenTime_hrs_Before", y="Academic_Performance_Before", data=df, label="Before", alpha=0.6)
sns.scatterplot(x="Daily_ScreenTime_hrs_After", y="Academic_Performance_After", data=df, label="After", alpha=0.6)
plt.title("Screen Time vs Academic Performance")
plt.xlabel("Screen Time (hrs)")
plt.ylabel("Academic Score")
plt.legend()
plt.show()

# BMI Before vs After
plt.figure(figsize=(7,5))
sns.kdeplot(df["BMI_Before"], label="Before", fill=True)
sns.kdeplot(df["BMI_After"], label="After", fill=True)
plt.title("BMI Changes Before vs After Intervention")
plt.xlabel("BMI")
plt.legend()
plt.show()

# Pairplot for deeper insight
sns.pairplot(df[["Daily_ScreenTime_hrs_Before", "Sleep_Hours_Before", "Academic_Performance_Before", "BMI_Before"]])
plt.suptitle("Pairwise Relationships (Before Intervention)", y=1.02)
plt.show()
```
